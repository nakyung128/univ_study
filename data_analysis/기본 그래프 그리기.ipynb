```python
# 8월의 최고 기온 데이터 시각화하기
import matplotlib.pyplot as plt
import csv

f = open('seoul.csv', encoding='UTF-8')
data = csv.reader(f)
next(data)
result=[]

for row in data:
    if row[-1] != '': # 최고 기온 값이 존재한다면
        if row[0].split('-')[1] == '08': # 8월에 해당하는 값이라면
            result.append(float(row[-1])) # result 리스트에 최고 기온 값 추가
plt.plot(result, 'hotpink') # result 리스트에 저장된 값을 hotpink 색으로 그리기
plt.show()
```

```python
# 내 생일의 최고 기온 데이터 시각화하기
import matplotlib.pyplot as plt
import csv

f = open('seoul.csv', encoding='UTF-8')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '01' and row[0].split('-')[2] == '28':
            result.append(float(row[-1]))
plt.plot(result, 'hotpink')
plt.show()
```

```python
# 내 생일의 최고 기온 및 최저 기온 데이터 시각화하기
import matplotlib.pyplot as plt
import csv

f = open('seoul.csv', encoding='UTF-8')
data = csv.reader(f)
next(data)
high = [] # 최고 기온 값을 저장할 리스트
low = [] # 최저 기온 값을 저장할 리스트

for row in data:
    if row[-1] != '' and row[-2] != '': # 최고 기온값과 최저 기온값이 존재한다면
        date = row[0].split('-') # 날짜값을 - 문자를 기준으로 구분해 저장
        if date[1] == '01' and date[2] == '28': # 1월 28일이면
            high.append(float(row[-1])) # 최고 기온 리스트에 추가
            low.append(float(row[-2])) # 최저 기온 리스트에 추가
            
plt.rc('font', family='Malgun Gothic') # 맑은 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.title('내 생일의 기온 변화 그래프') # 제목 설정
plt.plot(high, 'r', label='high') # 최고 기온 그래프
plt.plot(low, 'blue', label='low') # 최저 기온 그래프
plt.legend() # 범례 표시
plt.show() # 그래프 나타내기
```

```python
# 최고 기온 데이터 히스토그램으로 표현하기
import matplotlib.pyplot as plt
import csv

f = open('seoul.csv', encoding='UTF-8')
data = csv.reader(f)
next(data)
result=[]

for row in data:
    if row[-1] != '':
        result.append(float(row[-1]))
plt.hist(result, bins=100, color='red') # 히스토그램으로 나타내기
plt.show()
```

```python
# 8월의 최고 기온 히스토그램으로 표현하기
import matplotlib.pyplot as plt
import csv
f = open('seoul.csv', encoding='UTF--8')
data = csv.reader(f)
next(data)
result = [] # 8월의 최고 기온을 저장할 리스트 생성

for row in data:
    if row[-1] != '':
        month = row[0].split('-')[1] # '-'로 구분된 값 중 두 번째 값을 month에 저장
        if month == '08': # 8월달이라면
            result.append(float(row[-1])) # 리스트에 최고 기온값 추가
plt.hist(result, bins=100, color='r')
plt.show()
```

```python
# 1월과 8월의 최고 기온 데이터 히스토그램으로 표현하기
import matplotlib.pyplot as plt
import csv

f = open('seoul.csv', encoding='UTF-8')
data = csv.reader(f)
next(data)
jan=[] # 1월의 최고 기온값을 담을 리스트
aug=[] # 8월의 최고 기온값을 담을 리스트

for row in data:
    if row[-1] != '':
        month = row[0].split('-')[1]
        if month == '01': # 1월달이라면
            jan.append(float(row[-1]))
        if month == '08': # 8월달이라면
            aug.append(float(row[-1]))
plt.hist(jan, bins=100, color='blue', label='Jan') # 1월 히스토그램 생성
plt.hist(aug, bins=100, color='red', label='Aug') # 8월 히스토그램 생성
plt.legend() # 범례 표시
plt.show()
```

# 기온 데이터를 상자 그림으로 표현하기

```python
# 최고 기온 데이터 상자 그림으로 표현하기
import csv

f = open('seoul.csv', encoding='UTF-8')
data = csv.reader(f)
next(data)
result=[]

for row in data:
    if row[-1] != '':
        result.append(float(row[-1]))

import matplotlib.pyplot as plt
plt.boxplot(result) # 상자 그림으로 나타내기
plt.show()
```

```python
# 1월과 8월 최고 기온 데이터 상자 그림으로 표현하기
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='UTF-8')
data = csv.reader(f)
next(data)
aug = []
jan = []

for row in data:
    if row[-1] != '':
        month = row[0].split('-')[1] 
        if month == '08': # 8월이라면
            aug.append(float(row[-1]))
        if month == '01': # 1월이라면
            jan.append(float(row[-1]))
plt.boxplot([aug, jan])
plt.show()
```

```python
# 1월부터 12월까지 최고 기온 데이터 상자 그림으로 표현하기
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='UTF-8')
data = csv.reader(f)
next(data)

#월별 데이터를 저장할 리스트 month 생성 (12게)
month = [[],[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data:
    if row[-1] != '':
        # 월과 같은 번호의 인덱스에 월별 데이터 저장
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))
        
plt.boxplot(month)
plt.show()
```

```python
# 8월 1일부터 31일까지 최고 기온 데이터 상자 그림으로 표현하기
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='UTF-8')
data = csv.reader(f)
next(data)
day = [[] for i in range(31)]

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08':
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))
            
plt.style.use('ggplot')
plt.figure(figsize=(10, 5), dpi = 300)
plt.boxplot(day, showfliers = False)
plt.show()
```
