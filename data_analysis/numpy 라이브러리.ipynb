```python
import matplotlib.pyplot as plt
t = []
p2 = []
p3 = []
for i in range(0, 50, 2):
    t.append(i / 10)
    p2.append((i / 10) ** 2)
    p3.append((i / 10) ** 3)
plt.plot(t, t, 'r--', t, p2, 'bs', t, p3, 'g^')
plt.show()
```

```python
# numpy를 활용해 작성한 코드
import matplotlib.pyplot as plt
import numpy as np
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
# Mask 적용하기 (기본 데이터 만들기)
import numpy as np
a = np.arange(-5, 5)
print(a)
```

```python
# Mask 만들기
print(a<0)
```

```python
# Mask 적용하기
print(a[a<0]) # True인 값만 가져오기
```

```python
# Mask 적용하기2
mask1 = abs(a) > 3 # 절댓값이 3보다 큰 값
print(a[mask1])
```

```python
# 마스크 연결해서 사용하기
mask1 = abs(a) > 3 # 절댓값 > 3인 값
mask2 = abs(a) % 2 == 0 # 절댓값이 2의 배수
print(a[mask1+mask2]) # 둘 중 하나의 조건이라도 참일 경우 (or)
print(a[mask1*mask2]) # 두 가지 조건이 모두 참일 경우 (and)
```
