# pandas로 인구 구조 분석하기

```python
# 데이터 읽어오기

import pandas as pd
df = pd.read_csv('age.csv', encoding='utf-8', index_col=0)
df.head()
```

```python
# 전체 데이터 비율로 변환 후 총 인구 수, 연령 구간 인구수 열 삭제하기

import pandas as pd
df = pd.read_csv('age.csv', encoding='utf-8', index_col=0)
df = df.div(df['2021년10월_계_총인구수'], axis=0) # 1. 전체 데이터를 총인구수로 나눠서 비율로 변환
del df['2021년10월_계_총인구수'], df['2021년10월_계_연령구간인구수'] # 2. 총인구수, 연령구간인구수 열 삭제
name = input('원하는 지역의 이름을 입력해 주세요: ') # 3. 지역 이름 입력
a = df.index.str.contains(name) # 4. 해당 행을 찾아서 해당 지역의 인구 구조를 저장
df2 = df[a]
df2
```

```python
# 그래프로 시각화하기

import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
df2.T.plot()
plt.show()
```

```python
# 궁금한 지역(A)의 인구 구조 비율과 차이 계산하기

import numpy as np
# 궁금한 지역 A의 인구 비율에서 B의 인구 비율을 뺀다
x = df.sub(df2.iloc[0], axis=1)
# A의 인구 비율에서 B의 인구 비율을 뺀 값의 제곱 값을 모두 더한다
y = np.power(x, 2)
z = y.sum(axis=1)

# 가장 차이가 적은 상위 5개 지역 시각화하기
i = z.sort_values().index[:5] # 그 차이가 가장 작은 지역 5곳을 찾는다
df.loc[i].T.plot() # 결과를 꺾은선 그래프로 보여준다
plt.show()
```

```python
# 전체 코드

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)
df = pd.read_csv('age.csv', encoding='utf-8', index_col=0)
df = df.div(df['2021년10월_계_총인구수'], axis=0)
del df['2021년10월_계_총인구수'], df['2021년10월_계_연령구간인구수']

name = input('원하는 지역의 이름을 입력해 주세요: ')
a = df.index.str.contains(name)
df2 = df[a]

df.loc[np.power(df.sub(df2.iloc[0], axis=1), 2).sum(axis=1).sort_values().index[:5]].T.plot()

plt.show()
```
